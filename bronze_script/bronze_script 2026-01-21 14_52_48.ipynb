**### ## Creating Autoloader to get data and store it in bronze layer**

```python
## Getting and loading the Crm data in bronze layer
for name in crm_data:
    crm_info = (
        spark.read.option("header",True)
        .option("inferSchema",True)
        .csv("/Volumes/engineering/bronze/raw_data_sources/source_crm/"+name)
    )

    crm_info.write.mode("overwrite").saveAsTable("engineering.bronze."+name.replace(".csv",""))
    
```

```python
## Getting and loading the erp data in bronze layer
for name in erp_data:
    erp_info = (
        spark.read.option("header",True)
        .option("inferSchema",True)
        .csv("/Volumes/engineering/bronze/raw_data_sources/source_erp/"+name)
    )

    erp_info.write.mode("overwrite").saveAsTable("engineering.bronze."+name.replace(".csv",""))
```
